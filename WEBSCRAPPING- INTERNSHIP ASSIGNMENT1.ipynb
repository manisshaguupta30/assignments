# WEB SCRAPPING of "QUOTES TO SCRAPE" WEBSITE

WEB SCRAPPING:
Web Scraping is the process of extracting and parsing data from websites in an automated fashion using a computer program. There are many examples where web scraping can be useful, e.g. to create a job searching scraper, top movies scraper, mobile phone's price scraper and so on. Example, here we are extracting quotes. Web scraping involves extracting out information often from HTML documents and then convert into structured data like CSV file.

```python
#importing necessary libraries:
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://quotes.toscrape.com/page/1/"
```

```python
#Requests.get function will download the web page and returns page contents.

response = requests.get(url)
```

```python
html = response.content
```

```python
#To extract information from the HTML source code using programming, 
#we will use the Beautiful Soup library.
soup = BeautifulSoup (html , "html.parser")
soup
```

```python
#Upon inspecting the box containing information for the quotes,
#we will find a div tag with class as quote.

quotes = soup.find_all('div', {'class': 'quote'})
```

```python
#we can iterate through our quotes object and extract the text of all quotes.

for i in quotes:
    print((i.find('span', {'class':'text'})).text)
```

```python
#Again, we can use the div class “quote” to retrieve the data about the authors.
#furthermore, its data is inside small tag and class is author.

for i in soup.findAll("div",{"class": "quote"}):
    print((i.find("small", {"class": "author"})).text)
```

```python
#Information about the tags is available in the class “tags”.
#We need to extract the “content” from “meta”.

for i in soup.findAll("div",{"class": "tags"}):
    print((i.find("meta"))['content'])
```

```python
#Now we will create empty lists for all 10 description in first page and append the data into it

quotes = []
authors = []
tags = []

# loop over page 1 to 11
for pages in range(1,11): 
        
        url = "https://quotes.toscrape.com/page/1/"
        response = requests.get(url)
        html = response.content
        soup = BeautifulSoup (html , "html.parser")
        
        #quotes = soup.find_all('div',class_='quote')

        for i in soup.findAll("div",{"class":"quote"}):
                 quotes.append((i.find('span', {'class':'text'})).text)  
   
        for j in soup.findAll("div",{"class": "quote"}):
                 authors.append((j.find("small", {"class": "author"})).text)  
        
        for i in soup.findAll("div",{"class": "tags"}):
                 tags.append((i.find("meta"))['content'])
```

```python
#creating dataframe

df = pd.DataFrame(
    {'Quotes':quotes,
     'Authors':authors,
     'Tags':tags
    })
```

```python
#saving the file in csv format

df.to_csv('quote.csv', index = None)
```
